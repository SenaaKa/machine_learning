```python
def doldurucu(x):
    liste = x.columns
    for i in liste:
        if x[i].dtype == 'O':
            x[i]= x[i].fillna(x[i].mode()[0])
        else:
            x[i] = x[i].fillna(x[i].mean())
    return x 
```
