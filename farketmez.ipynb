```python
import pandas as pd
```

```python
df = pd.read_csv('employees.csv')
```

```python
df.info()
```

```python
df['First Name']=df['First Name'].fillna(df['First Name'].mode([0]))
```

```python
df.Gender=df.Gender.fillna(df.Gender.mode()[0])
```

```python
df['Senior Management']=df['Senior Management'].fillna(df['Senior Management'].mode()[0])
```

```python
df.Team=df.Team.fillna(df.Team.mode()[0])
```

```python
df.info()
```

```python
df
```

```python
df=df.drop(columns=['First Name', 'Last Login Time', 'Start Date'])
```

```python
df=pd.get_dummies(df)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor()
```

```python
df
```

```python
y=df.Salary
x=df.drop(columns=['Salary'])
```

```python
x_train=x.head(500)
x_test=x.tail(500)
y_train=y.head(500)
y_test=y.tail(500)
```

```python
dt.fit(x_train, y_train)
```

```python
y_predict=dt.predict(x_test)
```

```python
sena=df[df.Team_Distribution==5]
```
